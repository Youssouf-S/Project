```python
import tkinter as tk
from tkinter import filedialog, messagebox

class CalculatorApp(tk.Tk):
    def __init__(self):
        super().__init__()
        self.title("Calculatrice et Sauvegarde")

        # Frame principal
        self.main_frame = tk.Frame(self)
        self.main_frame.pack(padx=10, pady=10)

        # Boutons de sélection de fichier
        self.create_file_select_widgets()

        # Frame pour afficher les résultats
        self.result_frame = tk.Text(self.main_frame, height=5, width=50)
        self.result_frame.grid(row=2, column=0, columnspan=3, pady=10)  # Utilisation de grid
        #self.result_frame.pack(pady=10)  # Cette ligne est remplacée par la ligne ci-dessus

        # Boutons de calcul
        self.create_calc_buttons()

        # Champ de saisie pour le format de sauvegarde
        self.save_format_entry = tk.Entry(self.main_frame)
        self.save_format_entry.insert(0, "Entrez le format de sauvegarde (e.g., txt, json)")
        self.save_format_entry.grid(row=3, column=0, columnspan=3, pady=5)  # Utilisation de grid

        # Bouton de sauvegarde
        self.save_button = tk.Button(self.main_frame, text="Sauvegarder", command=self.save_results)
        self.save_button.grid(row=4, column=0, columnspan=3, pady=5)  # Utilisation de grid

    def create_file_select_widgets(self):
        self.file_label_1 = tk.Label(self.main_frame, text="Fichier 1:")
        self.file_label_1.grid(row=0, column=0, sticky="e")

        self.file_path_1 = tk.Text(self.main_frame, height=1, width=40)
        self.file_path_1.grid(row=0, column=1, padx=5)

        self.choose_file_button_1 = tk.Button(self.main_frame, text="Choisir Fichier 1", command=lambda: self.choose_file(1))
        self.choose_file_button_1.grid(row=0, column=2)

        self.file_label_2 = tk.Label(self.main_frame, text="Fichier 2:")
        self.file_label_2.grid(row=1, column=0, sticky="e")

        self.file_path_2 = tk.Text(self.main_frame, height=1, width=40)
        self.file_path_2.grid(row=1, column=1, padx=5)

        self.choose_file_button_2 = tk.Button(self.main_frame, text="Choisir Fichier 2", command=lambda: self.choose_file(2))
        self.choose_file_button_2.grid(row=1, column=2)

    def create_calc_buttons(self):
        self.add_button = tk.Button(self.main_frame, text="Additionner", command=self.add_files)
        self.add_button.grid(row=2, column=0, pady=5)

        self.subtract_button = tk.Button(self.main_frame, text="Soustraire", command=self.subtract_files)
        self.subtract_button.grid(row=2, column=1, pady=5)

    def choose_file(self, file_num):
        file_path = filedialog.askopenfilename()
        if file_path:
            if file_num == 1:
                self.file_path_1.delete("1.0", tk.END)
                self.file_path_1.insert(tk.END, file_path)
            elif file_num == 2:
                self.file_path_2.delete("1.0", tk.END)
                self.file_path_2.insert(tk.END, file_path)

    def add_files(self):
        file_path_1 = self.file_path_1.get("1.0", tk.END).strip()
        file_path_2 = self.file_path_2.get("1.0", tk.END).strip()
        # Ajouter votre logique de calcul ici
        result = f"Résultat de l'addition des fichiers: {file_path_1} + {file_path_2}"
        self.result_frame.delete("1.0", tk.END)
        self.result_frame.insert(tk.END, result)

    def subtract_files(self):
        file_path_1 = self.file_path_1.get("1.0", tk.END).strip()
        file_path_2 = self.file_path_2.get("1.0", tk.END).strip()
        # Ajouter votre logique de calcul ici
        result = f"Résultat de la soustraction des fichiers: {file_path_1} - {file_path_2}"
        self.result_frame.delete("1.0", tk.END)
        self.result_frame.insert(tk.END, result)

    def save_results(self):
        format = self.save_format_entry.get().strip()
        result = self.result_frame.get("1.0", tk.END).strip()
        if not format:
            messagebox.showerror("Erreur", "Veuillez spécifier le format de sauvegarde.")
            return
        # Ajouter votre logique de sauvegarde ici
        save_path = f"resultats.{format}"
        with open(save_path, 'w') as file:
            file.write(result)
        messagebox.showinfo("Sauvegarde", f"Résultats sauvegardés sous: {save_path}")


if __name__ == "__main__":
    app = CalculatorApp()
    app.mainloop()
```

```python
import numpy as np
import tkinter as tk
from tkinter import *
from tkinter import font, filedialog, messagebox
import json
from metricsyso import *


class JsonFile:
    def __init__(self, filepath):
        self.filepath = filepath
    
    def read_json(self):
        with open(self.filepath) as file:
            return json.load(file)

    def convdict(self):
        with open(self.filepath) as file:
            f = json.load(file)
        return f['conv_dict']
    
    def convclass(self):
        with open(self.filepath) as file:
            f = json.load(file)
        return f['conv_class']
    
    def conv_to_Truclass(self, conv_dict, conv_class):
        default_dict = {"1": "default", "2": "ground", "5": "vegetation", 
                        "6": "building", "7": "wall", 
                        "196": "structure", "211": "cable"
                        }
        model = {k: int(l) for k, v in conv_class.items() for l, m in default_dict.items() if v == m}
        return {k: j for k, v in conv_dict.items() for i, j in model.items() if int(i) == v }


def browse_file(title):
    filepath = filedialog.askopenfilename(title=title, filetypes=(("Laz or Las files", "*.laz *.las"), ("All files", "*.*")))
    return filepath


def compute_metrics(canvas, gtruth_file, pred_file, json_file, display_conf_matrix, display_weighted_metrics, display_metrics):
    op = JsonFile(json_file)
    conv_dict = op.convdict()
    conv_class = op.convclass()
    truth_class = op.conv_to_Truclass(conv_dict, conv_class)

    gtruth = read_file_class(gtruth_file)
    prediction = read_file_class(pred_file)
    count_pred = num_points(prediction)
    count_gtruth = num_points(gtruth)

    convertor = truth_class
    gtruth_conv = [convertor.get(str(k)) if convertor.get(str(k)) is not None else k for k in count_gtruth]

    input_pred = np.array(count_pred)
    input_gtruth = np.array(gtruth_conv)

    metrics_result = None
    if display_conf_matrix:
        metrics_result = compute_confusion_matrix(input_gtruth, input_pred)
    elif display_weighted_metrics:
        m = Metrics(input_gtruth, input_pred)
        metrics_result = m.weighted_metrics()
    elif display_metrics:
        m = Metrics(input_gtruth, input_pred)
        fr = m.compute_metrics()
        metrics_result =m.metrics_frame(fr) 

    # Update canvas with metrics result
    canvas.delete("all")
    if metrics_result:
        canvas.create_text(100, 100, text=metrics_result, anchor="nw", font=("Arial", 12))


def update_display():
    display_conf_matrix = conf_matrix_var.get()
    display_weighted_metrics = weighted_metrics_var.get()
    display_metrics = metrics_var.get()

    if display_conf_matrix:
        conf_matrix_canvas.pack()
    else:
        conf_matrix_canvas.pack_forget()

    if display_weighted_metrics:
        weighted_metrics_canvas.pack()
    else:
        weighted_metrics_canvas.pack_forget()

    if display_metrics:
        metrics_canvas.pack()
    else:
        metrics_canvas.pack_forget()


windows = Tk()
windows.geometry('900x500')
windows.title('Metrics')
windows['bg'] = 'grey'
windows.resizable(height=True, width=True)

bold_font = font.Font(weight="bold")

# Create variables to track user selections
conf_matrix_var = tk.BooleanVar(value=True)
weighted_metrics_var = tk.BooleanVar(value=True)
metrics_var = tk.BooleanVar(value=True)

# Create three canvas widgets
conf_matrix_canvas = tk.Canvas(windows, bg="white", width=300, height=200)
weighted_metrics_canvas = tk.Canvas(windows, bg="white", width=300, height=200)
metrics_canvas = tk.Canvas(windows, bg="white", width=300, height=200)

# Create buttons to toggle display
conf_matrix_checkbtn = tk.Checkbutton(windows, text="Confusion Matrix", variable=conf_matrix_var, command=update_display)
weighted_metrics_checkbtn = tk.Checkbutton(windows, text="Weighted Metrics", variable=weighted_metrics_var, command=update_display)
metrics_checkbtn = tk.Checkbutton(windows, text="Metrics", variable=metrics_var, command=update_display)

conf_matrix_checkbtn.pack()
weighted_metrics_checkbtn.pack()
metrics_checkbtn.pack()

conf_matrix_canvas.pack()
weighted_metrics_canvas.pack()
metrics_canvas.pack()

# Create buttons to load files
gtruth_file_btn = tk.Button(windows, text="Load Ground Truth File", command=lambda: gtruth_file_entry.insert(tk.END, browse_file("Select Ground Truth File")))
gtruth_file_btn.pack()
gtruth_file_entry = Entry(windows, width=50)
gtruth_file_entry.pack()

pred_file_btn = tk.Button(windows, text="Load Prediction File", command=lambda: pred_file_entry.insert(tk.END, browse_file("Select Prediction File")))
pred_file_btn.pack()
pred_file_entry = Entry(windows, width=50)
pred_file_entry.pack()

json_file_btn = tk.Button(windows, text="Load JSON File", command=lambda: json_file_entry.insert(tk.END, browse_file("Select JSON File")))
json_file_btn.pack()
json_file_entry = Entry(windows, width=50)
json_file_entry.pack()

# Create buttons to compute metrics
compute_button = tk.Button(windows, text="Compute Metrics",
                           command=lambda: compute_metrics(metrics_canvas, gtruth_file_entry.get(), pred_file_entry.get(), json_file_entry.get(),
                                                           conf_matrix_var.get(), weighted_metrics_var.get(), metrics_var.get()))
compute_button.pack()

windows.mainloop()
```
