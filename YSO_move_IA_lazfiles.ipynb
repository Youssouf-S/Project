```python
import os
import shutil

def trier_fichiers(source_folder, destination_folder, noms_specifies):
    # Créer le dossier de destination s'il n'existe pas
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Parcourir tous les fichiers du dossier source
    for filename in os.listdir(source_folder):
        # Vérifier si le nom du fichier correspond à l'un des noms spécifiés
        if any(name in filename for name in noms_specifies):
            # Construire le chemin complet du fichier source et de destination
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)
            # Déplacer le fichier vers le dossier de destination
            shutil.copy(source_path, destination_path)
            print(f"Fichier {filename} déplacé vers {destination_folder}")


```

```python
import os
import shutil

def trier_fichiers(source_folder, destination_folder, noms_specifies):
    # Créer le dossier de destination s'il n'existe pas
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Parcourir tous les fichiers du dossier source
    for filename in os.listdir(source_folder):
        # Vérifier si le nom du fichier correspond à l'un des noms spécifiés
        if any(name in filename for name in noms_specifies):
            # Construire le chemin complet du fichier source et de destination
            source_path = os.path.join(source_folder, filename)
            destination_path = os.path.join(destination_folder, filename)
            # Déplacer le fichier vers le dossier de destination
            shutil.copy(source_path, destination_path)
            print(f"Fichier {filename} déplacé vers {destination_folder}")


```

```python
import numpy as np
import laspy
import matplotlib.pyplot as plt 

class ClassPointsCounter:
    
    # specifing the path of the input file to extracte classes informations
    def __init__(self,file_path):
        self.file_path=file_path
        
    ## Return  a dictionnary of classes and numbers of points by class
    def count_points_per_class(self):
        class_counts = {}
        with laspy.open(self.file_path) as fh:
            las = fh.read()
            unique_classes, class_counts = np.unique(las.classification, return_counts=True)
            class_counts = dict(zip(unique_classes, class_counts))
        return class_counts
    
    # return only the number of points
    def number_points(self):
        class_counts=self.count_points_per_class()
        return class_counts.values()
    
    #return differents classes
    def class_names(self):
        class_counts=self.count_points_per_class()
        return class_counts.keys()
    
    # bar plot of number of points/class
    def plot_class_histogram(self):
        class_counts=self.count_points_per_class()
        x = list(class_counts.keys())
        y = list(class_counts.values())
        plt.bar(x, y,width =0.5, color = 'blue',edgecolor = 'black', linewidth = 2)
        plt.xlabel('Classe')
        plt.ylabel('Nombre de points')
        plt.title('Nombre de points par classe')
        plt.legend()
        plt.show()
        
        
    
            
```
